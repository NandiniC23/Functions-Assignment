#**1. What is the difference between a function and a method in Python?**

| Feature               | Function                        | Method                                           |
| --------------------- | ------------------------------- | ------------------------------------------------ |

        # Output: Hello, Nandini!

#**4. What is the purpose of the `return` statement in a Python function?**

-   The return statement in Python serves two main purposes:

1. It Ends the Function Execution
When the Python interpreter encounters a return statement, it immediately exits the function, even if there is more code below it.

This allows the function to stop running as soon as a result is ready.

Think of return as: “I'm done. Here’s the result.”

2. It Sends a Value Back to the Caller
The return statement specifies the value that the function should pass back to the code that called it.

**Example:**

def add(x, y):

  return x + y

result = add(3, 4)

print(result)  #Output: 7

| Aspect                     | Description                                 |
| -------------------------- | ------------------------------------------- |
| Ends function              | Stops further execution inside the function |
| Sends back a value         | Provides output to the calling code         |
| Optional                   | If not used, function returns `None`        |
| Can return multiple values | Returns as tuple                            |


#**5. What are iterators in Python and how do they differ from iterables?**

-   An iterable is any object you can loop over using a for loop. It implements the __iter__() method, which returns an iterator.

**Examples of iterables:**

Lists ([1, 2, 3])

Tuples ((1, 2))

Strings ("hello")

Sets, dictionaries

Any custom object with __iter__()

-  An iterator is an object that remembers its state as you traverse through it. It implements:

__iter__() → returns the iterator object itself

__next__() → returns the next item in the sequence or raises StopIteration


| Feature          | Iterable                       | Iterator                       |
| ---------------- | ------------------------------ | ------------------------------ |
| Definition       | Object that can be looped over | Object that produces next item |
| Has `__iter__()` |  Yes                          |Yes                          |
| Has `__next__()` |  No (usually)                 | Yes                          |
| Can be reused    |  Yes                          | No (one-time use)            |
| Example          | List, string, dict             | Object from `iter()`           |

my_list = [10, 20, 30]      # This is an iterable

my_iter = iter(my_list)     # This is an iterator

print(next(my_iter))        # 10

print(next(my_iter))        # 20

# print(next(my_iter))      # Raises StopIteration

Iterable: Can be looped over.

Iterator: Does the looping one item at a time using next().

#**6. Explain the concept of generators in Python and how they are defined.**

-   A generator is a special type of iterator that allows you to generate values on the fly, rather than storing them all in memory.

Generators are:

Defined using functions

Use the **yield** keyword instead of return

Lazy-evaluated: they produce items only when requested

How Generators Work:
1. yield Keyword:
The yield keyword is the key to generators. When a generator function encounters yield, it pauses its execution, remembers its state, and returns the yielded value.
2. Lazy Evaluation:
The generator function's code is not executed until you iterate over the generator object (e.g., using a for loop or next()).
3. Resuming Execution:
When you iterate again, the generator resumes execution from where it left off, picking up after the last yield statement.
4. State Preservation:
The generator preserves its state (local variables, etc.) between yield statements, allowing it to continue generating values based on its previous state.
5. Iterators:
Generators are iterators. They can be used in any context where an iterator is expected, such as for loops, in operator, etc.

**Example:**
def even_numbers(max_val):
    n = 0
    while n <= max_val:
        yield n
        n += 2

for num in even_numbers(10):
    print(num)  # Output: 0, 2, 4, 6, 8, 10

#**7. What are the advantages of using generators over regular functions?**

-  Generators in Python offer several advantages over regular functions, especially when dealing with large data sets or streaming data. Here are the key advantages:

1. Memory Efficiency
2. Lazy Evaluation
3. Infinite Sequences
4. Improved Performance
5. Cleaner Code with yield
6. Composability


**Example:**
# Generator
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1
# Regular function
def count_up_to(n):
    return list(range(1, n+1))

| Feature           | Generator                | Regular Function           |
| ----------------- | ------------------------ | -------------------------- |
| Memory Usage      | Low (yields one item)    | High (returns full result) |
| Evaluation        | Lazy (on-demand)         | Eager (immediate result)   |
| Infinite Sequence | Yes                      | No                         |
| Performance       | Often faster for streams | Slower for large datasets  |
| Code Complexity   | Lower for iterators      | May require more code      |

#**9. Explain the purpose and usage of the `map()` function in Python.**

-  The map() function in Python is used to apply a function to every item in an iterable (like a list or tuple) and returns a map object (an iterator).

| Use Case                  | Example                           |
| ------------------------- | --------------------------------- |
| Convert strings to ints   | `map(int, ["1", "2", "3"])`       |
| Apply formatting to text  | `map(str.upper, ["a", "b", "c"])` |
| Combine or transform data | `map(lambda x: x*2, [1, 2, 3])`   |

**Why Use map()?**
Cleaner and more readable than a for loop in simple transformations

Efficient - returns an iterator (doesn't compute everything at once)

Works well with lambda functions for concise code

| Feature      | Description                                    |
| ------------ | ---------------------------------------------- |
| Purpose      | Apply a function to each item in an iterable   |
| Returns      | A `map` object (iterator)                      |
| Common with  | `lambda`, list conversion, data transformation |
| Alternatives | List comprehensions, `for` loops               |

**Example**

def square(x):
    return x * x

numbers = [1, 2, 3, 4]
result = map(square, numbers)

print(list(result))  # Output: [1, 4, 9, 16]

#**10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**

-  These are three functional programming tools in Python that operate on iterables like lists or tuples.

**Example:**
Visual Example:

Given the list [1, 2, 3, 4]:

map(lambda x: x*2, [1, 2, 3, 4]) → [2, 4, 6, 8]

filter(lambda x: x%2 == 0, [1, 2, 3, 4]) → [2, 4]

reduce(lambda x, y: x + y, [1, 2, 3, 4]) → 1

| Function   | Purpose        | Input                      | Output               | Use Case                                |
| ---------- | -------------- | -------------------------- | -------------------- | --------------------------------------- |
| `map()`    | Transformation | Function + Iterable        | Transformed iterator | Squaring numbers, formatting text       |
| `filter()` | Selection      | Predicate + Iterable       | Filtered iterator    | Filtering even numbers, removing blanks |
| `reduce()` | Aggregation    | Binary Function + Iterable | Single value         | Sum, product, or combining all items    |

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200

def is_prime(n):
    if n < 2 or n > 200:
        return False  # Outside the valid range
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True
print(is_prime(7))
print(is_prime(58))
print(is_prime(125))
print(is_prime(203))
```

```python
#9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

# List of temperatures in Celsius
celsius_temps = [0, 20, 37, 100]

# Function to convert Celsius to Fahrenheit
def c_to_f(c):
    return (c * 9/5) + 32

# Use map to apply the conversion to each temperature
fahrenheit_temps = list(map(c_to_f, celsius_temps))

print(fahrenheit_temps)
```

```python
#10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

def remove_vowels(s):
    vowels = "aeiouAEIOU"
    # Filter out characters that are vowels
    return ''.join(filter(lambda char: char not in vowels, s))

# Example usage
input_string = "Hello, Nandini?"
result = remove_vowels(input_string)
print(result)
```
